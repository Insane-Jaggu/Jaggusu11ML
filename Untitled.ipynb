```python
import tweepy 
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re 
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
dataset = pd.read_csv('AnimeList.csv')
dataset
```

```python
data2 = dataset[dataset['sentiment']==2]
data2
```

```python
allWords = ' '.join([twts for twts in data2['tweet']])
wordCloud = WordCloud(width=500, max_font_size=500, max_words=250, background_color="white", margin=10).generate(allWords)
plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show()
wordCloud.to_file('Image.png')
print('Image is saved successfully')
```
